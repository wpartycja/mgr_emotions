```python
import matplotlib.pyplot as plt

class_counts = {
    "anger": 1103,
    "disgust": 2,
    "excited": 1041,
    "fear": 40,
    "frustration": 1849,
    "happiness": 595,
    "neutral": 1708,
    "sadness": 1084,
    "surprise": 107,
}

labels = list(class_counts.keys())
counts = list(class_counts.values())

plt.figure(figsize=(10, 6))
bars = plt.bar(labels, counts, color="pink", edgecolor="black")

plt.title("IEMOCAP emotion distribution")
plt.xlabel("Emotion")
plt.ylabel("Count")
plt.xticks(rotation=45)

for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20,
             str(count), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
class_counts_meld = {
    "anger": 1607,
    "disgust": 361,
    "fear": 358,
    "joy": 2308,
    "neutral": 6434,
    "sadness": 1002,
    "surprise": 1636,
}

labels = list(class_counts_meld.keys())
counts = list(class_counts_meld.values())

plt.figure(figsize=(10, 6))
bars = plt.bar(labels, counts, color="pink", edgecolor="black")

plt.title("MELD emotion distribution")
plt.xlabel("Emotion")
plt.ylabel("Count")
plt.xticks(rotation=45)

for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
             str(count), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
iemocap4cls_counts = {
    "angry": 1103,
    "happy": 1636,
    "neutral": 1708,
    "sad": 1084
}

meld4cls_counts = {
    "angry": 1607,
    "happy": 2308,
    "neutral": 6434,
    "sad": 1002
}

labels = list(iemocap4cls_counts.keys())
iemocap_values = list(iemocap4cls_counts.values())
meld_values = list(meld4cls_counts.values())

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

bars1 = axes[0].bar(labels, iemocap_values, color="pink", edgecolor="black")
axes[0].set_title("IEMOCAP (4 classes)")
axes[0].set_xlabel("Emotion")
axes[0].set_ylabel("Count")
axes[0].set_xticks(range(len(labels)))
axes[0].set_xticklabels(labels, rotation=45)

for bar, count in zip(bars1, iemocap_values):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20,
                 str(count), ha='center', va='bottom', fontsize=9)

bars2 = axes[1].bar(labels, meld_values, color="pink", edgecolor="black")
axes[1].set_title("MELD (4 classes)")
axes[1].set_xlabel("Emotion")
axes[1].set_xticks(range(len(labels)))
axes[1].set_xticklabels(labels, rotation=45)

for bar, count in zip(bars2, meld_values):
    axes[1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20,
                 str(count), ha='center', va='bottom', fontsize=9)

plt.suptitle("Emotion distribution comparison: IEMOCAP vs MELD", fontsize=14, y=1.05)
plt.tight_layout()
plt.show()

```
